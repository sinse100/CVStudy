```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

### Image Histogram 에 대해 알아봅시다.
+ ROI 에 대한 Image Histogram 을 구해봅시다.

```python
img = cv2.imread('img/mountain.jpg',cv2.IMREAD_GRAYSCALE) 

x,y,w,h = cv2.selectROI('img',img,False,False)            # ROI 선택
cv2.destroyAllWindows()

new_img = np.zeros(img.shape[:2],np.uint8) 
new_img[y:y+h,x:x+w] = 255                                # Image 의 ROI 영역의 모든 Channel 값들을, 255 로 변경 

hist = cv2.calcHist([img],[0],new_img,[256],[0,256])      # ROI 에 대해서만 Histogram 생성

## cv2.calcHist
##  => img : Input Video 로, 각각의 Frame 들을 원소로 갖는 List
##  => channel : Histogram 계산 대상이 될 Channel 로, 첫번째 Channel부터, 0, 1, 2, ... 의 번호가 부여되며, 해당 번호들을 원소로 갖는 List
##  => mask : Image 와 Shape이 같고, Histogram을 구하고자 하는 영역의 Channel 값들은 모두 255(그외는 0) 로 지정되어있는 ndarray 
##  => histSize : 계급의 개수로, 각 Channel 이 가질 수 있는 값의 개수들을 원소로 갖는 List
##  => ranges : Channel 값의 범위를 List 로 표현한 것 [min, max+1]
##  => return -> 각 계급에 속하는 Channel 의 개수들 (np.ndarray)

plt.plot(hist)              # 명암 분포를 Plot 에 표현

print(type(hist))
print(hist.shape)          
print(hist.sum())
plt.show()
```

+ RGB 컬러 스페이스의 Image Histogram 을 분석해봅시다. (Red, Blue, Green Channel 별)

```python
img = cv2.imread('img/mountain.jpg')                        

blue_hist = cv2.calcHist([img],[0],None,[256],[0,256])              # Blue Channel 에 대한 Histogram 계산
green_hist = cv2.calcHist([img],[1],None,[256],[0,256])             # Green Channel 에 대한 Histogram 계산
red_hist = cv2.calcHist([img],[2],None,[256],[0,256])               # Red Channel 에 대한 Histogram 계산

plt.subplot(111)
plt.plot(blue_hist , color='b')
plt.plot(green_hist, color='g') 
plt.plot(red_hist, color='r')

plt.show()

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image/영상 에서의 정규화/표준화에 대해 알아봅시다.
+ 정규화 (Normalization)<br> : 어떤 범위 내에 분포하는 데이터들을, 원하는 다른 범위내에 분포하도록하되, 기존, 데이터들 간의 차이 정보(해당 Data Set 의 특성, 가령, 분포의 모양)들은 그대로 유지시키는 것 
  + 기존 데이터 셋의 최소, 최대 값을 $Min, Max$ 라고 하고,정규화하고자 하는 새로운 범위를 $(newMin, newMax)$ 라고 할때, 정규화 이전의 각각의 Data $I$를 정규화한 후의 값$I_{new}$ 는 다음의 공식을 통해 유도 가능
  $$ I_{new} = newMin + (I-Min) \frac{(newMax-newMin)}{(Max-Min)} $$<br> 위의 공식에서, $\frac{(newMax-newMin)}{(Max-Min)}$ 는 기존 분포에서의 1 이라는 데이터 값 간 차이가, 새로운 분포 내에서는 얼마가 되는지를 의미.<br><br>
  + 구간을 사용한 정규화의 단점<br>: <b>이상치 (Outlier) 문제</b> 를 해결할 수 없음. 여기서, 이상치 문제란, 간단히 말해 다른 데이터와 비교하여 차이가 큰 데이터 (이를 점 이상치라고 함)를 중점적으로 의미. 
  
  > 만일, 100개의 값이 있는데 그 중 99개는 0과 40 사이에 있고, 나머지 하나가 100이면 어떨까. 그러면 99개의 값이 모두 0부터 0.4 사이의 값으로 변환되고, 1 개의 값(기존 100 값)만 1로 변환되고, 데이터가 양 극에 치우치는 문제가 발생한다.
     + 특히, 머신러닝 분야에서, 인공신경망의 가중치 갱신에, 입력값의 변화율 또한 그 영향을 미치게 되므로, 특정 Feature 에 점 이상치가 존재하는 경우, 가중치의 갱신 정도에 해당 점 이상치가 큰 영향을 줄 것이고, 이는, 다른 Feature 보다, 해당 점 이상치가 존재하는 Feature 에 Model 이 과적합되는 현상을 일으킬 수 있음<br><br>
+ 표준화 (Standardization, Z-Score Normalization)<br>: 데이터가 존재할 수 있는 상한과 하한의 개념을 정해두지 않고, 데이터들을 최대한 고르게 퍼뜨리는 것(정확히는 기존의 데이터 분포를 표준정규분포로 변환하는 것)에 초점을 둔 것으로, 이상치 문제를 완벽히 해결할 수는 없으나, 상대적으로 덜 민감하다는 특징을 갖고 있음
  + 기존 분포에서의 데이터 포인트의 값을 $x_{i}$, 평균을 $m$, 표준편차를 $\sigma$ 라고 했을 때, 각 데이터 포인트들에 대한 $Z-Score_{i}$는 아래와 같음
  $$Z-Score_{i} = \frac{x_{i} - m}{\sigma}$$
    + 이는, 각 데이터 포인트들에 평균 $m$ 을 먼저 빼, 데이터 포인트들을 평행이동 (평균은 0 이 됨)시키고, 각 데이터 포인트들에 표준편차 $\sigma$ 를 나눠, 표준편차를 1 (평균 $m$ 을 뺀 시점에서, 평균을 제외한 대부분의 데이터 포인트들은, 평균적으로 $0 \pm \sigma$ 에 위치할 것임) 로 만들어줌
    + 당연히, 평균으로부터 특히 멀리 떨어져 있는 값(점 이상치)이 많을 수록, 표준편차가 커지게 되고 이에 따라 점 이상치들은  표준화 후, 평균 0 에 더 가까워질 것이다. 이로써 구간 정규화보다, 조금 더 데이터의 분포가 균일해짐. (그림 (나)-2 참조)
  + <b>주의!</b>
데이터를 z-score로 변환한다고 해서 실제 해당 데이터가 정규분포를 이루게되는 것은 아니다. 단지 비교(평균 0을 기준으로, 각각의 데이터 포인트들이 얼마나 퍼져있는가)를 목적으로 표준정규분포와 같은 척도로 만드는 것 뿐이다. 따라서, 표준화시킨 후의 데이터에서, 해당 표준정규 분포에서 추출할 수 있는 의미있는 데이터(데이터의 기존 분포에서 왜곡되지 않는 것)는 <b>평균으로부터의 거리의 대소 관계(or, 각 데이터 포인트들 간, 평균으로부터 떨어진 정도의 비율)</b>라고 할 수 있음.

    + 표준화 후, 상대적으로 이상치와 평균간의 거리가 눈에 띄게 줄어들음.

#### 명암 구분이 뚜렷하지 않은 이미지의 화질을, 구간 정규화를 사용하여 개선해봅시다.

```python
img = cv2.imread('img/abnormal.jpg', cv2.IMREAD_GRAYSCALE)

## Numpy 이용한 정규화
img_f = img.astype(np.float32)
print(f'Before Norm : min → {img_f.min()}, max → {img_f.max()}')
img_norm = (img_f - img_f.min()) * (255 / (img_f.max() - img_f.min())) 
img_norm = img_norm.astype(np.uint8)
dst_img = np.zeros_like(img,dtype=np.uint8)
print(dst_img)
## cv2.normalize 이용한 정규화
img_norm2 = cv2.normalize(img,None,0,255,cv2.NORM_MINMAX)
## cv2.normalize
##  => img : 정규화하고자 하는 대상 이미지 (ndarray)
##  => dst_img : 정규화 된 최종 결과 이미지를 저장할 ndarray (img 와 shape 같아야 함)
##  => min : -> 구간 정규화 : 구간의 최솟값
##           -> Norm 사용 정규화 : Norm 값
##  => max : -> 구간 정규화 : 구간의 최댓값 
##  => algo_flag : 정규화 알고리즘을 지정
##                 -> cv2.NORM_MINMAX : 구간 정규화 

hist = cv2.calcHist([img],[0],None,[256],[0,255])
hist_norm = cv2.calcHist([img_norm],[0],None,[256],[0,255])
hist_norm2 = cv2.calcHist([img_norm2],[0],None,[256],[0,255])
```

```python
cv2.imshow('Before', img)
cv2.imshow('Manual', img_norm)
cv2.imshow('cv2.normalize()',img_norm2)

hists = {'Before' : hist, 'Manual' : hist_norm, 'cv2.normalize()' : hist_norm2}
for i,(k,v) in enumerate(hists.items()):
    plt.subplot(1,3,i+1)
    plt.title(k)
    plt.plot(v)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
